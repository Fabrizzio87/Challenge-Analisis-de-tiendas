### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1_df = pd.read_csv(url)
tienda2_df = pd.read_csv(url2)
tienda3_df = pd.read_csv(url3)
tienda4_df = pd.read_csv(url4)

tienda1_df.head()
```

#1. Análisis de facturación



```python

ingreso_tienda1 = tienda1_df["Precio"].sum()
ingreso_tienda2 = tienda2_df["Precio"].sum()
ingreso_tienda3 = tienda3_df["Precio"].sum()
ingreso_tienda4 = tienda4_df["Precio"].sum()

print(f"Ingreso total de tienda1: {ingreso_tienda1}")
print(f"Ingreso total de tienda2: {ingreso_tienda2}")
print(f"Ingreso total de tienda3: {ingreso_tienda3}")
print(f"Ingreso total de tienda4: {ingreso_tienda4}")
```

# 2. Ventas por categoría

```python
ventas_por_categoria_tienda1 = tienda1_df.groupby('Categoría del Producto').size().reset_index(name='Cantidad de Ventas')
ventas_por_categoria_tienda2 = tienda2_df.groupby('Categoría del Producto').size().reset_index(name='Cantidad de Ventas')
ventas_por_categoria_tienda3 = tienda3_df.groupby('Categoría del Producto').size().reset_index(name='Cantidad de Ventas')
ventas_por_categoria_tienda4 = tienda4_df.groupby('Categoría del Producto').size().reset_index(name='Cantidad de Ventas')


print("\nCantidad de productos vendidos por categoría en la Tienda 1:")
print(ventas_por_categoria_tienda1)
print("\n")

print("\nCantidad de productos vendidos por categoría en la Tienda 2:")
print(ventas_por_categoria_tienda2)
print("\n")

print("\nCantidad de productos vendidos por categoría en la Tienda 3:")
print(ventas_por_categoria_tienda3)
print("\n")

print("\nCantidad de productos vendidos por categoría en la Tienda 4:")
print(ventas_por_categoria_tienda4)
print("\n")
```

# 7. Análisis Tiendas

Análisis de la sucursal menos eficiente: Tienda 4

Para el análisis se tomaron tres métricas, ingresos totales por tienda, calificación promedio por tienda y costo de envío por tienda.

1. Ingresos Totales: Representan la cantidad total de dinero generado por las ventas.
2. Calificación Promedio de los Clientes: Indica el nivel de satisfacción del cliente con la tienda.
3. Costo de Envío Promedio: Mide los gastos asociados con la entrega de los productos.

Tienda 4 tiene los ingresos más bajos, a pesar de vender casi la misma cantidad de productos que las demás tiendas. Esto muestra que probablemente está vendiendo productos más baratos o con menor margen de ganancia. Sin embargo, tiene el costo de envío más bajo y una calificación aceptable (4.00)


Conclusión:

Dado que los ingresos son un indicador clave de eficiencia en un contexto de ventas, y considerando que la Tienda 4 tiene los ingresos más bajos, la Tienda 4 es la menos eficiente.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # Para gráficos más avanzados y heatmaps

# --- 1. Preparación de los Datos ---

# Añadir una columna para identificar la tienda antes de combinar
tienda1_df['Tienda_ID'] = 'Tienda 1'
tienda2_df['Tienda_ID'] = 'Tienda 2'
tienda3_df['Tienda_ID'] = 'Tienda 3'
tienda4_df['Tienda_ID'] = 'Tienda 4'

# Combinar todos los DataFrames
all_tiendas_df = pd.concat([tienda1_df, tienda2_df, tienda3_df, tienda4_df], ignore_index=True)

# Verificar tipos de datos (deberían ser float64 para lat, lon, Precio, Calificación)
print("Información del DataFrame consolidado:")
all_tiendas_df.info()
print("\n")

# --- 2. Mapeo de Ventas por Tienda (Gráficos de Dispersión) ---

# Gráfico de dispersión consolidado
plt.figure(figsize=(12, 8))
sns.scatterplot(data=all_tiendas_df, x='lon', y='lat', hue='Tienda_ID', palette='viridis', s=50, alpha=0.7)
plt.title('Distribución Geográfica de Ventas por Tienda')
plt.xlabel('Longitud')
plt.ylabel('Latitud')
plt.legend(title='Tienda')
plt.grid(True)
plt.show()

# Gráficos de dispersión individuales (opcional, si quieres ver cada tienda por separado)
fig, axes = plt.subplots(2, 2, figsize=(15, 12), sharex=True, sharey=True)
plt.suptitle("Distribución Geográfica de Ventas - Individual por Tienda", fontsize=16, y=1.02)
tiendas_dfs = {'Tienda 1': tienda1_df, 'Tienda 2': tienda2_df, 'Tienda 3': tienda3_df, 'Tienda 4': tienda4_df}
colors = {'Tienda 1': 'blue', 'Tienda 2': 'green', 'Tienda 3': 'red', 'Tienda 4': 'purple'}

for i, (nombre_tienda, df_tienda) in enumerate(tiendas_dfs.items()):
    ax = axes[i//2, i%2]
    ax.scatter(df_tienda['lon'], df_tienda['lat'], alpha=0.6, s=30, color=colors[nombre_tienda])
    ax.set_title(nombre_tienda)
    ax.set_xlabel('Longitud')
    ax.set_ylabel('Latitud')
    ax.grid(True)

plt.tight_layout(rect=[0, 0, 1, 1]) # Ajustar para que el suptitle no se superponga
plt.show()

# --- 3. Identificación de Áreas con Mayor Concentración de Ventas ---

# Mapa de calor (Heatmap) usando KDE (Kernel Density Estimate) con Seaborn
plt.figure(figsize=(12, 8))
sns.kdeplot(data=all_tiendas_df, x='lon', y='lat', fill=True, thresh=0.05, cmap='viridis', levels=100)
plt.title('Mapa de Calor de Concentración de Ventas (Todas las Tiendas)')
plt.xlabel('Longitud')
plt.ylabel('Latitud')
plt.grid(True)
plt.show()

# También podrías hacer heatmaps por tienda si es necesario:
# for nombre_tienda, df_tienda in tiendas_dfs.items():
# plt.figure(figsize=(10, 7))
#     sns.kdeplot(data=df_tienda, x='lon', y='lat', fill=True, thresh=0.05, cmap='viridis', levels=100)
# plt.title(f'Mapa de Calor de Concentración de Ventas - {nombre_tienda}')
# plt.xlabel('Longitud')
# plt.ylabel('Latitud')
# plt.grid(True)
# plt.show()

# --- 4. Análisis de Rendimiento Geográfico ---

# a) Ingresos por ubicación
plt.figure(figsize=(14, 10))
scatter_ingresos = sns.scatterplot(
    data=all_tiendas_df,
    x='lon',
    y='lat',
    hue='Tienda_ID',
    size='Precio',  # El tamaño del punto representa el precio/ingreso
    sizes=(20, 500), # Rango de tamaños para los puntos
    alpha=0.7,
    palette='magma'
)
plt.title('Distribución Geográfica de Ingresos por Venta')
plt.xlabel('Longitud')
plt.ylabel('Latitud')
plt.legend(title='Tienda e Ingreso', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()

# b) Calificaciones por ubicación
plt.figure(figsize=(14, 10))
scatter_calificaciones = sns.scatterplot(
    data=all_tiendas_df,
    x='lon',
    y='lat',
    hue='Tienda_ID',
    palette='coolwarm',
    size='Calificación', # El tamaño del punto representa la calificación (o puedes usar 'c' para color continuo)
    # Opcional: usar 'c' para un mapeo de color continuo para la calificación
    # c=all_tiendas_df['Calificación'], cmap='coolwarm_r',
    sizes=(30,300),
    alpha=0.7
)
# Si usaste 'c' para calificación, necesitarás una colorbar:
# norm = plt.Normalize(all_tiendas_df['Calificación'].min(), all_tiendas_df['Calificación'].max())
# sm = plt.cm.ScalarMappable(cmap="coolwarm_r", norm=norm)
# sm.set_array([])
# scatter_calificaciones.figure.colorbar(sm, label='Calificación')

plt.title('Distribución Geográfica de Calificaciones de Clientes')
plt.xlabel('Longitud')
plt.ylabel('Latitud')
plt.legend(title='Tienda y Calificación', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()

# c) Explorar rendimiento por 'Lugar de Compra' (agregado)
# Esta parte es más analítica que puramente visual en un mapa,
# pero puede complementar el análisis geográfico.

ingresos_por_lugar = all_tiendas_df.groupby(['Lugar de Compra', 'Tienda_ID'])['Precio'].agg(['sum', 'mean', 'count']).reset_index()
calificacion_por_lugar = all_tiendas_df.groupby(['Lugar de Compra', 'Tienda_ID'])['Calificación'].agg(['mean', 'count']).reset_index()

print("\nIngresos Agregados por Lugar de Compra y Tienda:")
print(ingresos_por_lugar.sort_values(by='sum', ascending=False).head(10))

print("\nCalificación Promedio por Lugar de Compra y Tienda:")
print(calificacion_por_lugar.sort_values(by='mean', ascending=False).head(10))

# Podrías crear gráficos de barras comparando tiendas dentro de los lugares de compra más importantes.
# Ejemplo para los 5 lugares con más ventas totales:
top_lugares_ventas = all_tiendas_df['Lugar de Compra'].value_counts().nlargest(5).index
df_top_lugares = all_tiendas_df[all_tiendas_df['Lugar de Compra'].isin(top_lugares_ventas)]

plt.figure(figsize=(15, 7))
sns.barplot(data=df_top_lugares, x='Lugar de Compra', y='Precio', hue='Tienda_ID', estimator=sum, errorbar=None)
plt.title('Ingresos Totales por Tienda en los 5 Lugares con Más Ventas')
plt.ylabel('Ingresos Totales')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

plt.figure(figsize=(15, 7))
sns.barplot(data=df_top_lugares, x='Lugar de Compra', y='Calificación', hue='Tienda_ID', estimator='mean', errorbar=None) # estimator='mean' es el por defecto para barplot con y numérico
plt.title('Calificación Promedio por Tienda en los 5 Lugares con Más Ventas')
plt.ylabel('Calificación Promedio')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
